```python
def get_language(city):
    if city=='Chennai' or city=='Podicherry':
        return 'Tamil'
    elif city=='Mumbai' or city=='Delhi' or city=='Noida':
        return 'Hindi'
    elif city=='Calcutta':
        return 'Bengali'
    elif city=='Cochin':
        return 'Malayalam'
    elif city=='Hyderabad':
        return 'Telugu'
    elif city=='Bangalore':
        return 'Kannadam'
    else:
        return 'ENGLISH'
    
df['Language']=df['City'].apply(get_language)
```
